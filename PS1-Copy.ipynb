```python
%%sql with B as (Select A.i as i, A.j as j, A.val as val
                from A)
SELECT A.i as i, B.j as j, SUM(A.val * B.val) as val
        FROM A, B
        WHERE A.j = B.i
        GROUP BY A.i, B.j;
```

```python
%%sql with D as (with C as (
            with B as (Select A.i as i, A.j as j, A.val as val from A)
            SELECT A.i as i, B.j as j, SUM(A.val * B.val) as val
            FROM A, B
            WHERE A.j = B.i
            GROUP BY A.i, B.j)
    Select A.i as i, C.j as j, SUM(A.val * C.val) as val
    from A, C
    where A.j = C.i
    group by A.i, C.j)
SELECT A.i as i, D.j as j, SUM(A.val * D.val) as val
FROM A, D
WHERE A.j = D.i 
GROUP BY A.i, D.j;
```

```python
%%sql with H as (select WeekDate from Holidays 
            where Holidays.IsHoliday = 'TRUE'),
HolidaySales as (select Store, WeeklySales, WeekDate 
                 from Sales 
                 where WeekDate in H),
TotalStoreSales as (select Store, SUM(WeeklySales) as total
                    from HolidaySales
                    group by Store)
select Store, MAX(total) as AllSales
from TotalStoreSales
group by Store;
```

```python
%%sql with TotalStoreSales as (select Store, SUM(WeeklySales) as 'total'
                    from Sales,Holidays 
                    where Sales.WeekDate = Holidays.WeekDate
                    and Holidays.IsHoliday = 'TRUE'
                    group by Store)
select Store, MAX(total) as AllSales
from TotalStoreSales;
```

```python
%%sql with NonHolidays as (select SUM(s.WeeklySales) as SumWeeklySales, s.WeekDate
                          from Holidays h, Sales s
                          where h.WeekDate = s.WeekDate 
                          and h.IsHoliday = 'FALSE'
                          group by s.WeekDate),
HolidaySales as (select SUM(s.WeeklySales) as SumWeeklySales, s.WeekDate
                          from Holidays h, Sales s
                          where h.WeekDate = s.WeekDate 
                          and h.IsHoliday = 'TRUE'
                          group by s.WeekDate)
Select COUNT(*) as NumNonHolidays
from NonHolidays
where SumWeeklySales > (Select AVG(SumWeeklySales) 
                     from HolidaySales);
```

```python
%%sql select st.Type as type, SUM(s.WeeklySales) as TotalSales
from Stores st, Sales s
where s.Store = st.Store 
and (s.WeekDate LIKE '%-06-%' or s.WeekDate LIKE '%-07-%' or s.WeekDate LIKE '%-08-%')
group by st.Type;
```

```python
%%sql select S3.B as company, (S1.d+S2.d+S3.d) as distance
from streets S1, streets S2, streets S3
where S1.A = 'UW-Madison' and S2.A = S1.B and S3.A = S2.B and distance <= 9
UNION
select S2.B as company, (S1.d+S2.d) as distance
from streets S1, streets S2
where S1.A = 'UW-Madison' and S2.A = S1.B and distance <= 9
UNION
select B as company, d
from streets 
where A = 'UW-Madison' and d <= 9;
```

```python
%%sql select S1.A as company_1, S2.B as company_2, (S1.d+S2.d) as distance
from streets S1, streets S2
where S2.A = 'UW-Madison' and S2.A = S1.B and distance <=15
and company_1 < company_2
UNION
select S1.A as company_1, S3.B as company_2, (S1.d+S2.d+S3.d) as distance
from streets S1, streets S2, streets S3
where S2.A = S1.B and S2.A = 'UW-Madison' and S3.A=S2.B and distance <=15
and company_1 < company_2
UNION
select S1.A as company_1, S3.B as company_2, (S1.d+S2.d+S3.d) as distance
from streets S1, streets S2, streets S3
where S2.A = S1.B and S3.A = 'UW-Madison' and S3.A=S2.B and distance <=15
and company_1 < company_2
UNION
select S1.A as company_1, S4.B as company_2, (S1.d+S2.d+S3.d+S4.d) as distance
from streets S1, streets S2, streets S3, streets S4
where S2.B = 'UW-Madison' and S2.A = S1.B and S3.A=S2.B and S4.A=S3.B and distance <=15
and company_1 < company_2;
```
